## Handling free input

```python
from nltk.stem.snowball import SnowballStemmer
from nltk.tokenize import word_tokenize, sent_tokenize
from sklearn.feature_extraction.text import CountVectorizer
import time

# input:
#     col: columns that containing free text
#     k: top k important terms(features) will be consider
#     output: n x k matrix indicating features
        
def auto_stemming(col, k):
    print("start stemming")
    start = time.time()
    i = 0 
    stemmed_data = []
    for message in col:
        if message != message: 
            stemmed_data.append(' ')
            continue
        a = []
        for sent in sent_tokenize(message):
            for word in word_tokenize(sent):
                a.append(SnowballStemmer("english", ignore_stopwords=True).stem(word))
        stemmed_data.append(' '.join(a))
        i += 1
    print (time.time()-start)
    print(len(stemmed_data))
    print('stemming finished\n')
    
    print("start vectorizing")
    vectorizer = CountVectorizer()
    X = vectorizer.fit_transform(stemmed_data)
    topindex = np.argsort(X.toarray().sum(axis=0))[::-1][:k]
    topterms = [vectorizer.get_feature_names()[i] for i in topindex]
    top = X.toarray()[:,topindex]
    print("finish vectorizing")
    
    return topterms, top
```
